```python
import scipy.stats as stats
import xml.etree.ElementTree as ET
import datetime as dt
import pandas as pd
import re
```

```python
file_path = ("E:\python_project\extract-xbrl-financial\AmbujaCement_INDAS_790050_4966_22102022125650_WEB.xml")
tree = ET.parse(file_path)
root = tree.getroot()
```

```python
cols = []
for x in root:
    x.tag = re.sub("{http://www.bseindia.com/xbrl/fin/2020-03-31/in-bse-fin}","",x.tag)
    cols.append(x.tag)
```
